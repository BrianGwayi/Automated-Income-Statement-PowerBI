```python
import pandas as pd
import numpy as np


def bin_data(x):
    if x == 1:
        return 'very unsatisfied'
    elif x == 2:
        return 'unsatisfied'
    elif x == 3:
        return 'neutral'
    elif x == 4:
        return 'satisfied'
    else:
        return 'very satisfied'


url = "/content/CSATSurvey.csv"
x = pd.read_csv(url) #Read csv file

x = (
  x.drop(columns=['Email'], axis=1) # dropped column
  .rename({'Unnamed: 4':'Relationship'}, axis=1) # renamed column
  .assign(ClientID = lambda x: x.Date.str.split(" ").str[0]) # split & assign column
  .drop(columns=['Date'], axis=1) # dropped column
  .dropna(subset=['CSAT']) # dropped NaN
  .drop_duplicates(subset=['ClientID'], keep='first') # dropped duplicates
  .query("CSAT <= 5") # filter outliers
  #.convert_dtypes(infer_objects=True)
  .astype({'CSAT':'int64', 'Region':'category',
           'Relationship':'category','Survey date':'datetime64[ns]'}) # type conversion
  .assign(Categories = x.CSAT.apply(bin_data)) # assign & apply categories
  )
```
